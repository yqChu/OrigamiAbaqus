```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Displacement and Velocity

peak_disp_mag = np.zeros((10, 2))
peak_vel_mag = np.zeros((10, 2))
index = 0

for amp_id in range(10):
    # plt.figure(figsize=(8,3))
    for load_type in ["Left", "Right"]:
        for data_type in ["Disp", "Vel"]:
            data = np.loadtxt(f"C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3-CreaseStiffeness-f=200-AmplitudeSweeping/{load_type}Load{data_type}Freq=200Amp={amp_id+1}hor_num=4.txt",skiprows=3)
            time = data[:,0]
            disp = data[:,1]
            magnitudes = np.fft.fft(disp)
            frequencies = np.fft.fftfreq(len(disp), 
            (time[-1]-time[0])/len(time))
            # if data_type=="Disp":
            #     plt.subplot(1,2,1)
            # else:
            #     plt.subplot(1,2,2)
            # if load_type=="Left":
            #     plt.plot(frequencies*2*np.pi, abs(magnitudes), 'r', label=f"{load_type}, {data_type}")
            # else:
            #     plt.plot(frequencies*2*np.pi, abs(magnitudes), 'b', label=f"{load_type}, {data_type}")
            # plt.title(f"Amplitude={amp_id+1}")
            # plt.grid(True)
            # plt.xlim(0,500)
            # plt.legend(loc='upper right')

            w_min = 0
            w_max = 500
            indices = np.where((frequencies>=w_min) & (frequencies*2*np.pi<=w_max))[0]
            selected_freqs = frequencies[indices]
            selected_mags = magnitudes[indices]
            peak_idx = np.argmax(abs(selected_mags))
            peak_freq = selected_freqs[peak_idx]
            if load_type=="Left":
                if data_type=="Disp":
                    peak_disp_mag[index,0] = abs(selected_mags[peak_idx])
                else:
                    peak_vel_mag[index,0] = abs(selected_mags[peak_idx])
            else:
                if data_type=="Disp":
                    peak_disp_mag[index,1] = abs(selected_mags[peak_idx])
                else:
                    peak_vel_mag[index,1] = abs(selected_mags[peak_idx])
    index += 1
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(range(1,11), peak_disp_mag[:,0], c='r', s=20, label="LeftLoad")
plt.scatter(range(1,11), peak_disp_mag[:,1], c='b', s=20, label="RightLoad")
plt.grid(True)
plt.title("Displacement FFT Magnitude -- Shaking Amplitude")
plt.xlabel("Amplitude")
plt.ylabel("Magnitude")
plt.legend(loc="lower right")

plt.subplot(1,2,2)
plt.scatter(range(1,11), peak_vel_mag[:,0], c='r', s=20, label="LeftLoad")
plt.scatter(range(1,11), peak_vel_mag[:,1], c='b', s=20, label="RightLoad")
plt.grid(True)
plt.title("Velocity FFT Magnitude -- Shaking Amplitude")
plt.xlabel("Amplitude")
plt.ylabel("Magnitude")
plt.legend(loc="lower right")
```
