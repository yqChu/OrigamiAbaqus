```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load 1 Left U2-1
Sinusoidal force loading on the left, report the y direction displacement of the 1st (all 6) left loading node.

```python
data = np.loadtxt('C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3/Load1LeftU2-1.txt',skiprows=3)
Load1LeftU2_1 = data[:,1]
T = data[:,0]

fft_result = np.fft.fft(Load1LeftU2_1)
frequencies = np.fft.fftfreq(len(Load1LeftU2_1), (T[-1]-T[0])/len(T))

plt.figure(figsize=(16,8))

plt.subplot(2,2,1)
plt.plot(T, Load1LeftU2_1)
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(frequencies, fft_result)
plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, 0.4)

data = np.loadtxt('C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3/Load1LeftV2-1.txt',skiprows=3)
Load1LeftV2_1 = data[:,1]
T = data[:,0]

fft_result = np.fft.fft(Load1LeftV2_1)
frequencies = np.fft.fftfreq(len(Load1LeftV2_1), (T[-1]-T[0])/len(T))

plt.subplot(2,2,3)
plt.plot(T, Load1LeftV2_1)
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(frequencies, fft_result)
plt.grid(True)
plt.xlim(0, 100)
```
