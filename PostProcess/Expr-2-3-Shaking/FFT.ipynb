```python
# Mesh quantitiy's influence on that unexpected peak

for Freq_id in range(100, 400, 100):
    data = np.loadtxt(f"C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3-Shaking/LeftLoadFreq={Freq_id}Disphor_num=3.txt",skiprows=3)
    Disp = data[:,1]
    Time_disp = data[:,0]
    fft_result_disp=np.fft.fft(Disp)
    frequencies_disp = np.fft.fftfreq(len(fft_result_disp), (Time_disp[-1]-Time_disp[0])/len(Time_disp))
    data = np.loadtxt(f"C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3-Shaking/LeftLoadFreq={Freq_id}Velhor_num=3.txt",skiprows=3)
    Vel = data[:,1]
    Time_vel = data[:,0]
    fft_result_vel=np.fft.fft(Vel)
    frequencies_vel = np.fft.fftfreq(len(fft_result_vel), (Time_vel[-1]-Time_vel[0])/len(Time_vel))
    plt.figure(figsize=(12,3))
    plt.subplot(1,4,1)
    plt.plot(Time_disp, Disp, 'r', label="mesh=3")
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,2)
    plt.plot(frequencies_disp*2*np.pi, abs(fft_result_disp), 'r', label="mesh=3")
    plt.xlim(0, 700)
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,3)
    plt.plot(Time_vel, Vel, 'r', label="mesh=3")
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,4)
    plt.plot(frequencies_vel*2*np.pi, abs(fft_result_vel), 'r', label="mesh=3")
    plt.xlim(0, 700)
    plt.legend(loc="upper right")
    plt.grid(True)
    del data, Disp, Time_disp, fft_result_disp, frequencies_disp
    del Vel, Time_vel, fft_result_vel, frequencies_vel

for Freq_id in range(100, 400, 100):
    data = np.loadtxt(f"C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3-Shaking/LeftLoadFreq={Freq_id}Disphor_num=5.txt",skiprows=3)
    Disp = data[:,1]
    Time_disp = data[:,0]
    fft_result_disp=np.fft.fft(Disp)
    frequencies_disp = np.fft.fftfreq(len(fft_result_disp), (Time_disp[-1]-Time_disp[0])/len(Time_disp))
    data = np.loadtxt(f"C:/Research/Nonreciprocity/OrigamiAbaqus/PostProcess/Expr-2-3-Shaking/LeftLoadFreq={Freq_id}Velhor_num=5.txt",skiprows=3)
    Vel = data[:,1]
    Time_vel = data[:,0]
    fft_result_vel=np.fft.fft(Vel)
    frequencies_vel = np.fft.fftfreq(len(fft_result_vel), (Time_vel[-1]-Time_vel[0])/len(Time_vel))
    plt.figure(figsize=(12,3))
    plt.subplot(1,4,1)
    plt.plot(Time_disp, Disp, 'r', label="mesh=5")
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,2)
    plt.plot(frequencies_disp*2*np.pi, abs(fft_result_disp), 'r', label="mesh=5")
    plt.xlim(0, 700)
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,3)
    plt.plot(Time_vel, Vel, 'r', label="mesh=5")
    plt.legend(loc="upper right")
    plt.grid(True)

    plt.subplot(1,4,4)
    plt.plot(frequencies_vel*2*np.pi, abs(fft_result_vel), 'r', label="mesh-5")
    plt.xlim(0, 700)
    plt.legend(loc="upper right")
    plt.grid(True)
    del data, Disp, Time_disp, fft_result_disp, frequencies_disp
    del Vel, Time_vel, fft_result_vel, frequencies_vel

```
